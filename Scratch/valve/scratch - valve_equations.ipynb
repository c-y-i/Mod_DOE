## Mechanics solutions relating valve 'spring pressure' to parameter designs

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Import constants and equations from 2024 van Laake

# from paper
Ks = 2.9e10 # N/m
Kr = 3.6e5 # N*m/rad

# estimation of our system
b0 = 3.75e-3 # m
h0 = 3e-3 # m

# discretize x
x = np.linspace(-h0, h0, 1000) # [h0]

# from paper
Av = np.pi*b0**2 # [b0]^2
ks_num = np.sqrt(h0**2 + b0**2) - np.sqrt(x**2 + b0**2) # [b0]
ks_den = np.sqrt(x**2 + b0**2) # [b0]
x_atan = [math.atan2(xx, b0) for xx in x] # [rad]
kr_num = math.atan2(h0,b0)*np.ones(len(x_atan)) + x_atan # [rad]
kr_den = b0*(x**2/b0**2 + 1) # [b0]
Fs = -Ks*ks_num/ks_den*x + Kr*kr_num/kr_den # N
Ps = Fs/Av # N/m^2
```

```python
b0**2
```

```python
plt.plot(x/b0, 'r')
plt.plot(x**2/b0**2, 'b')
```

```python
min_arg = np.argmin(np.abs(x))
```

```python
# plot x vs. Ps
plt.plot(x, Ps)
# star at x=0, horizontal line at Ps(0)
plt.plot(0, Ps[min_arg], 'r*', markersize=10)
plt.axhline(Ps[min_arg], color='k', linestyle='--')
plt.xlabel('x (m)')
plt.ylabel('Ps (N/m^2)')
plt.show()
print('Ps(0) =', Ps[min_arg]/1000, 'kPa') # N/m^2
```
